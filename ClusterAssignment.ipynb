Problem Statement: 
In this assignment students have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference. 


```
import numpy as np 
from sklearn.cluster import KMeans 
from scipy import misc
```

```
import matplotlib.pyplot as plt
import scipy.misc
face = scipy.misc.face(gray=True)
plt.imshow(face)
plt.show()
```

```
face.shape
```

```
rows = face.shape[0]
cols = face.shape[1]
```

```
image=face.reshape(rows*cols,1)
```

```
kmeans=KMeans(n_clusters=5)
```

```
kmeans.fit(image)
```

```
kmeans.cluster_centers_
```

```
labels=kmeans.labels_
```

```
labels = labels.reshape(rows,cols);
```

```
labels.shape
```

```
plt.imsave('compressed.png',labels);
```

```
image = plt.imread('compressed.png')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
axes[0].set_title("Compressed Image")
axes[0].imshow(image)
axes[1].set_title("Raw Image")
axes[1].imshow(face)
fig.tight_layout()
```
